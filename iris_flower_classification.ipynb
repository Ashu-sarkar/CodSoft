# Model Implementation

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
iris_data_updated['species'] = label_encoder.fit_transform(iris_data_updated['species'])

iris_data_updated['species'].unique()
```

```python
# Splitting Data
from sklearn.model_selection import train_test_split
x = iris_data_updated.drop(['species'], axis = 1) # independent variable

y = iris_data_updated['species'] # dependent variable
```

```python
x_sepal_train, x_sepal_test, y_species_train, y_species_test = train_test_split(x,y,random_state = 10, test_size=0.3)
```

```python
x_sepal_train, x_sepal_test, y_species_train, y_species_test.shape
```

```python
x_sepal_train
```

```python
# Applying Linear Regression Model

from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
# Feeding Data into Model

model.fit(x_sepal_train, y_species_train)
y_species_predicted = model.predict(x_sepal_test)
```

```python
# Accuracy of Linear Regression Model

from sklearn.metrics import accuracy_score

sc_lr = round(model.score(x_sepal_test, y_species_test) * 100, 2)
print("Accuracy", str(sc_lr), "%")
```

```python
# Applying Logistic Regression Model

from sklearn.linear_model import LogisticRegression
logistic_model = LogisticRegression(max_iter = 130)
logistic_model.fit(x_sepal_train,y_species_train)

LogisticRegression(max_iter= 130)

y_species_logistic_predicted = logistic_model.predict(x_sepal_test)
```
